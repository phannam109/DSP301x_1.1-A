# Task 2: Kiểm tra tính hợp lệ của dữ liệu

```python
# Xem thông tin chung về bảng dữ liệu
table.info()
```

```python
# Lọc bảng dữ liệu để lấy bảng dữ liệu hợp lệ:
valid_table1 = table[(table.index.str.len() == 9) & (table.index.str.startswith('N'))]
valid_table2 = valid_table1[valid_table1.index.str.removeprefix("N").str.isdigit()]
valid_table2
```

```python
# Lọc bảng dữ liệu để lấy bảng dữ liệu không hợp lệ
invalid_table2 = table[~table.index.isin(valid_table2.index)]
invalid_table2
```

```python
# Làm việc với bảng để đánh giá các câu trả lời
# Câu trả lời trống, điền giá trị 0 tương đương 0 điểm
valid_table3 = valid_table2.fillna(0)
valid_table3.head()
```

```python
# Đáp án đúng theo đầu bài
answer_key = ("B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D").split(",")
```

```python
for i in range(25): 
    valid_table3.loc[:,i + 1][valid_table3.loc[:,i+1] == answer_key[i]] = 4
    valid_table3.loc[:,i + 1][(valid_table3.loc[:,i+1] != 4) & (valid_table3.loc[:,i+1] != 0)] = -1
valid_table3
```

```python
valid_table3.info()
```

```python
valid_table3 = valid_table3.loc[:,1:25].astype(int)
valid_table3.info()
```

```python
valid_table3["Total"] = valid_table3.loc[:,1:25].sum(axis = 1)
```

```python
total_score = valid_table3["Total"]
total_score.head()
```
